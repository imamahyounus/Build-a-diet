```python
#Source: https://health.gov/dietaryguidelines/2015/guidelines/appendix-2/#males

import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "Calorie_intake.csv"
calories = pd.read_csv(csv_path, encoding='ISO-8859-1')
calories.head(100)
```

```python
plt.plot(calories['Age Group'], calories['Sedentary (Men)'], linewidth=3, color='midnightblue')
plt.plot(calories['Age Group'], calories['Moderate (Men)'], linewidth=3, color='blue')
plt.plot(calories['Age Group'], calories['Active (Men)'], linewidth=3, color='skyblue')
plt.legend(loc='center left', bbox_to_anchor=(1, .7))
plt.xticks(rotation=90)
plt.title(f"Recommended Calorie Intake Per Day by Age and Physical Activity Level (Men)")
plt.ylabel("Calories Per Day")
plt.xlabel("Age Group")
plt.show()
```

```python
plt.plot(calories['Age Group'], calories['Sedentary (Women)'], linewidth=3, color='purple')
plt.plot(calories['Age Group'], calories['Moderate (Women)'], linewidth=3, color='magenta')
plt.plot(calories['Age Group'], calories['Active (Women)'], linewidth=3, color='pink')
plt.legend(loc='center left', bbox_to_anchor=(1, .7))
plt.xticks(rotation=90)
plt.title(f"Recommended Calorie Intake Per Day by Age and Physical Activity Level (Women)")
plt.ylabel("Calories Per Day")
plt.xlabel("Age Group")
plt.show()
```
